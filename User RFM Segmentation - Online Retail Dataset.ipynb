# Data Cleaning

```python
df_clean = df.copy()
```

```python
# create date column
df_clean['date'] = pd.to_datetime(df_clean['order_date']).dt.date.astype('datetime64[ns]')
df_clean.info()
```

```python
# remove null value in customer_id column
df_clean = df_clean[~df_clean['customer_id'].isna()]
df_clean.info()
```

```python
# makes all product_name columns lowercase
df_clean['product_name'] = df_clean['product_name'].str.lower()
df_clean.head(3)
```

```python
# remove all rows with product_code or product_name 'test'
df_clean = df_clean[~(df_clean['product_code'].str.lower().str.contains('test')) | 
                    ~(df_clean['product_name'].str.lower().str.contains('test'))]

df_clean.head()
```

```python
# create an order_status column with the value 'canceled' if the order_id begins with the letter 'c' and 
# 'delivered' if the order_id does not begin with the letter 'c'
df_clean['order_status'] = np.where(df_clean['order_id'].str[:1] == 'C', 'cancelled', 'delivered')
df_clean.head(3)
```

```python
# change the negative quantity value to positive because the negative value only indicates 
# that the order is canceled
df_clean['quantity'] = df_clean['quantity'].abs()
df_clean.head(3)
```

```python
# creates an amount value column, which is the multiplication of quantity and price
df_clean['amount'] = df_clean['quantity'] * df_clean['price']
df_clean.head(3)
```

```python
# replace the product_name of a product_code that has multiple product_names 
# with one of its most frequently occurring product_names
most_freq_product_name = df_clean.groupby(['product_code','product_name'], as_index=False).agg(order_cnt=('order_id','nunique')).sort_values(['product_code','order_cnt'], ascending=[True,False])
most_freq_product_name['rank'] = most_freq_product_name.groupby('product_code')['order_cnt'].rank(method='first', ascending=False)
most_freq_product_name = most_freq_product_name[most_freq_product_name['rank']==1].drop(columns=['order_cnt','rank'])
df_clean = df_clean.merge(most_freq_product_name.rename(columns={'product_name':'most_freq_product_name'}), how='left', on='product_code')
df_clean['product_name'] = df_clean['most_freq_product_name']
df_clean = df_clean.drop(columns='most_freq_product_name')
df_clean.head(3)
```

```python
# convert customer_id column to string
df_clean['customer_id'] = df_clean['customer_id'].astype(str)
df_clean.info()
```

```python
# remove outliers
from scipy import stats
df_clean = df_clean[(np.abs(stats.zscore(df_clean[['quantity','amount']]))<3).all(axis=1)]
df_clean = df_clean.reset_index(drop=True)
df_clean.info()
```

# Create RFM Segmentation

```python
df_user = df_clean.groupby('customer_id', as_index = False).agg(order_cnt = ('order_id','nunique'),
                                                                max_order_date = ('date', 'max'),
                                                                total_order_value = ('amount', 'sum'))
df_user
```

Aggregate transaction data to create RFM columns, namely the last order date of each user (max_order_date), total transactions (order_cnt), and total transaction value (total_order_value).

```python
today = df_clean['date'].max()
df_user['day_since_last_order'] = (today - df_user['max_order_date']).dt.days
df_user.head()
```

```python
today
```

Create a column for the number of days since the customer placed their last order. We assume 'today' is the date of the last order in the dataset.

```python
df_user.describe()
```

```python
df_user['recency_score'] = pd.cut(df_user['day_since_last_order'],
                                 bins = [df_user['day_since_last_order'].min(),
                                        np.percentile(df_user['day_since_last_order'], 20),
                                        np.percentile(df_user['day_since_last_order'], 40),
                                        np.percentile(df_user['day_since_last_order'], 60),
                                        np.percentile(df_user['day_since_last_order'], 80),
                                        df_user['day_since_last_order'].max()],
                                 labels = [5, 4, 3, 2, 1],
                                 include_lowest = True).astype(int)
df_user.head()
```

Create a binning of the number of days since the last order consisting of 5 bins with boundaries, namely, Min, P20, P40, P60, P80, Max. Then, label 1 to 5 from the highest bin to the lowest as the recency score.

```python
df_user['frequency_score'] = pd.cut(df_user['order_cnt'],
                                 bins = [ 0, # df_user['order_cnt'].min(),<-- cannot be used because 
                                             # from the descriptive statistics the value of the 20th percentile 
                                             # (order_cnt) is only 1, so the lower limit cannot be 1 as well.
                                        np.percentile(df_user['order_cnt'], 20),
                                        np.percentile(df_user['order_cnt'], 40),
                                        np.percentile(df_user['order_cnt'], 60),
                                        np.percentile(df_user['order_cnt'], 80),
                                        df_user['order_cnt'].max()],
                                 labels = [1, 2, 3, 4, 5],
                                 include_lowest = True).astype(int)
df_user.head()
```

Create a binning of the total transactions (orders) consisting of 5 bins with boundaries, namely, Min, P20, P40, P60, P80, Max. Then, label 1 to 5 from the lowest bin to the highest as the frequency score.

```python
df_user['monetary_score'] = pd.cut(df_user['total_order_value'],
                                 bins = [df_user['total_order_value'].min(),
                                        np.percentile(df_user['total_order_value'], 20),
                                        np.percentile(df_user['total_order_value'], 40),
                                        np.percentile(df_user['total_order_value'], 60),
                                        np.percentile(df_user['total_order_value'], 80),
                                        df_user['total_order_value'].max()],
                                 labels = [1, 2, 3, 4, 5],
                                 include_lowest = True).astype(int)
df_user.head()
```

Create a binning of the total order value consisting of 5 bins with boundaries, namely, Min, P20, P40, P60, P80, Max. Then, label 1 to 5 from the lowest bin to the highest as the monetery score.

```python
df_user.info()
```

```python
df_user['segment'] = np.select(
       [(df_user['recency_score'] == 5) & (df_user['frequency_score'] >= 4),
        (df_user['recency_score'].between(3, 4)) & (df_user['frequency_score'] >= 4),
        (df_user['recency_score'] >= 4) & (df_user['frequency_score'].between(2, 3)),
        (df_user['recency_score'] <= 2) & (df_user['frequency_score'] == 5),
        (df_user['recency_score'] == 3) & (df_user['frequency_score'] == 3),
        (df_user['recency_score'] == 5) & (df_user['frequency_score'] == 1),
        (df_user['recency_score'] == 4) & (df_user['frequency_score'] ==1),
        (df_user['recency_score'] <=2) & (df_user['frequency_score'].between(3, 4)),
        (df_user['recency_score'] == 3) & (df_user['frequency_score'] <= 2),
        (df_user['recency_score'] <= 2) & (df_user['frequency_score'] <= 2)],
       ['01-Champion', '02-Loyal Customers', '03-Potential Loyalists', "04-Can't Lose Them", '05-No Need Attention',
       '06-New Customers', '07-Promising', '08-At Risk', '09-About to Sleep', '10-Hibernating'])
df_user
```

Create segment columns based on the recency and frequency scores.

```python
summary = pd.pivot_table(df_user, index = 'segment',
              values = ['customer_id', 'day_since_last_order', 'order_cnt', 'total_order_value'],
              aggfunc = {'customer_id' : pd.Series.nunique,
                        'day_since_last_order' : [np.mean, np.median],
                        'order_cnt' : [np.mean, np.median],
                        'total_order_value' : [np.mean, np.median]})
summary['pct_unique'] = (summary['customer_id'] / summary['customer_id'].sum() * 100).round(1) # calculating user percentage
summary
```

# Analysis Result

From the results of customer segmentation using the RFM method, there are several insights, including: 
1. Most users are in the Hibernating segment (1060 or 27.3% customers), Champion (550 or 14.1% customers), Loyal Customers (546 or 14% customers).
2. Special programs that focus on the urgency of transactions for Loyal Customers who have an average last transaction of 40 days ago, thus making them transact again in the near future and can move up to the Champion segment.
3. Special programs that focus on the number of transactions for Potential Loyalists who have an average number of orders of 2 times, thus making them transact more frequently and can move up to the Champion segment.
4. Special programs for the Hibernating segment can be carried out to make them return to transactions even though they are not so frequent, thus they can move up to the New Customers or even Potential Loyalists segment.
